<h1>TASK 1 - Image to Pencil Sketch with Python</h1>

<h4>Importing Libraries</h4>

```python
import cv2
from google.colab.patches import cv2_imshow
```

<h4>Reading Original Image</h4>

```python
image = cv2.imread("/virat kohli.jpg")
cv2_imshow(image)
```

<h4>Converting original image to grayscale</h4>

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

<h4>Inverting the grayscale image</h4>

```python
inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
```

<h4>Blurring the image using Gaussian function</h4>

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

<h3>Pencil Sketch</h3>

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
```

###Image - Pencil sketch

```python
cv2_imshow(image)
cv2_imshow(pencil_sketch)
```
